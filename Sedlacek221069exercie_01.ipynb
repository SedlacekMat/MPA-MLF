# Exercise 1.5-1:

The Fibonacci Sequence is a series of numbers. The next number is found by adding up the two numbers before it. The first two numbers are 0 and 1. For example, 0, 1, 1, 2, 3, 5, 8, 13. The next number in this series above is 8 + 13 = 21



Create a function that takes an integer as an input and return a list that contains all fibonaci numbers smaller than input integer

example:

*print(fibonaci(10))*

[0, 1, 1, 2, 3, 5, 8]


```python
def fibonaci(upper_threshold: int) -> list:
  seed = [0, 1]
  while seed[-1] < upper_threshold:
    seed.append(seed[-1] + seed[-2])
  return seed[0:-1]

print(fibonaci(100))
```

- Create a numpy array with shape 5x5, that contains number from 25..to 0
- create a function that takes an arbitary numpy array as an input and sets all numbers smaller than user defined threshold to 0
- test your function on your array
- make one implementation using loops and one withnout using loops, compare execution time with *time* library

```python
import numpy as np
import time

testMatrix = np.random.randint(0,25,(5,5))

# Using 2 for loops
def threshold_loop(testArray, threshold):
    numRows = int(len(testArray))
    numCols = int(np.size(testArray)/numRows)
    for row in range(numRows):
        for col in range(numCols):
            if testArray[row][col] < threshold:
                testArray[row][col] = 0
    print(testArray)

# Using mostly just indexing
def threshold_noloop(testArray, threshold):
    testArray[testArray<threshold] = 0
    print(testArray)

# The time library is largely inaccurate in Jupyter's online version, when measured offline the times are roughly equal
print(testMatrix,"\n\n")
tic1 = time.perf_counter()
threshold_loop(testMatrix,15)
toc1 = time.perf_counter()
print("\n\n")
tic2 = time.perf_counter()
threshold_noloop(testMatrix,15)
toc2 = time.perf_counter()
print("Loop time: {loop:.3f} ms; No-loop time {noloop:.3f} ms".format(loop=(toc1-tic1)*1e3, noloop=(toc2-tic2)*1e3))
```

```python
import matplotlib.pyplot as plt


def show_in_digi(number: int) -> None:
    symbols = {
        '0': ('111', '101', '101', '101', '111'),
        '1': ('001', '001', '001', '001', '001'),
        '2': ('111', '001', '111', '100', '111'),
        '3': ('111', '001', '111', '001', '111'),
        '4': ('101', '101', '111', '001', '001'),
        '5': ('111', '100', '111', '001', '111'),
        '6': ('111', '100', '111', '101', '111'),
        '7': ('111', '001', '001', '001', '001'),
        '8': ('111', '101', '111', '101', '111'),
        '9': ('111', '101', '111', '001', '111'),
        '.': ('000', '000', '000', '000', '001'),
    }

    digits = [symbols[digit] for digit in str(number)]
    out = []
    for i in range(5):
        temp = list("0"+"0".join(segment[i] for segment in digits)+"0")
        floatTemp = [float(x) for x in temp]
        out.append(floatTemp)

    plt.imshow(out, cmap='gray')
    plt.show()


show_in_digi(237965)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('DataCviko1.csv')
#dataset.describe() #displays main statistics of the dataset

#display rows where there's more than 310 bedrooms
lotsBedrooms = dataset[dataset['total_bedrooms'] > 310]
display(lotsBedrooms)

#no first/last row
noFirstLast = dataset.head(-1).tail(-1)
display(noFirstLast)

#households mean and plot
meanHouseholds = dataset['households'].mean()
plt.plot(dataset['households'],'bo')
plt.hlines(meanHouseholds,xmin=0,xmax=2999,color='r')
plt.show()

#checking for and replacing NaN with mean of ENTIRE dataset
print("Any non-number values?: ", dataset.isnull().values.any())
if dataset.isnull().values.any():
    dataset.fillna(dataset.mean())
    
#plot lat and long
latitude = dataset['latitude']
longitude = dataset['longitude']
plt.plot(latitude,longitude,'rx')
plt.show()

#minmax normalisation
minPar = dataset[['population','median_income']].min()
maxPar = dataset[['population','median_income']].max()
normData = (dataset[['population','median_income']]-minPar)/(maxPar-minPar)
display(normData)

#correlation
corrData = dataset.corr()
display(corrData)
```
